```python
import pandas as pd

data = {
    'Employee_ID': [101, 102, 103, 104, 102, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Bob', 'Eve'],
    'Age': [24, 21, 28, 35, 21, 19],
    'Salary': [50000, 30000, 80000, 120000, 30000, 45000]
}

# Convert that dictionary into a Pandas DataFrame.
df = pd.DataFrame(data)
print("Original data:")
print(df)

# The web scraper accidentally grabbed Bob twice! Write the code to drop the duplicate row.
print("\nAfter dropping duplicate:")
df_clean = df.drop_duplicates()
print(df_clean)

# Create a new filtered table (let's call it vip_df) that only includes employees who are 
# older than 22 AND have a salary strictly greater than 45,000.
vip_df = df_clean[(df_clean['Age']>22) & (df_clean['Salary'] > 45000)]

print("\nNew vip table:")
print(vip_df)
```
