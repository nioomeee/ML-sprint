```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

data = {
    'Income': [15000, 1800, 120000, 10000], 
    'Age': [25, 18, 42, 51]
}
df = pd.DataFrame(data)

print("\nOriginal data:")
print(df)
```

```python
scaler = MinMaxScaler()

# telling the scaler which columns to transform
# using double brackets because we're passinng a list of columns

df[['Income', 'Age']] = scaler.fit_transform(df[['Income', 'Age']])

print("\nPerfectly scaled data: ")
print(df)
```
