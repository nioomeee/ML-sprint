```python
import pandas as pd

data = {
    'Name': ['John', 'Anna', 'Peter', 'John'],
    'Age': [24, 13, 53, 24]
}

df = pd.DataFrame(data)

print("Original table: ")
print(df)
```

```python
# basic Duplicate
# by default, finds duplicate rows, keeps 1st one it sees, deletes the rest
df_basic = df.drop_duplicates()

print("\nBasic Clean(Kept 1st John): ")
print(df_basic)
```

```python
# Advanced Clean
# using 'subset' specifying column to check and also which occurence to keep

df_advanced = df.drop_duplicates(subset=['Name'], keep='last')

print("\n Advanced Clean(Kept last John): ")
print(df_advanced)
```
