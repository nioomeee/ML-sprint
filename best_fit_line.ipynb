```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = {
    'House_Size': [1000, 1500, 2000, 2500, 3000],
    'House_Price': [200000, 250000, 300000, 350000, 400000]
}
df = pd.DataFrame(data)

X = df[['House_Size']]
y = df['House_Price']

model = LinearRegression()

model.fit(X, y)

print("\n Model trained successfully!")

# plot data as dots (actual houses)
plt.scatter(X, y, color='purple', label='Actual Prices')

# Ai predicts values for all houses in X table
ai_predictions = model.predict(X)

plt.plot(X, ai_predictions, color = 'black', linewidth = 2, label = 'AI trend line')

plt.title("AI trend analysis")
plt.xlabel('Size (sq. ft.)')
plt.ylabel('Price ($)')
plt.show()
```
