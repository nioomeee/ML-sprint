```python
import pandas as pd
import numpy as np

data = {
    'Income': [1, 10, 100, 1000, 10000, 100000]
}

df = pd.DataFrame(data)

print(f"Messy skewed data:\n {df}")

# transformation 1 : square root
df['Sqrt_Income'] = np.sqrt(df['Income'])

# transformaation 2: log
df['Log_Income'] = np.log10(df['Income'])

print("\n Transformed data: ")
print(df)
```
