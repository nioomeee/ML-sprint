```python
import pandas as pd
data = {
    'Name': ['John', 'Anna', 'Peter', 'Linda'],
    'Age': [24, 13, 53, 33],
    'Salary': [50000, 2000, 100000, 30000]
}
df = pd.DataFrame(data)

print("Original data")
print(df)
```

```python
# Rule 1: Age >= 25 AND Age <= 60
# Rule 2: Salary > 10000

filtered_data = df[(df['Age'] >= 25) & (df['Age'] <= 60) & (df['Salary'] > 10000)]

print("\nFiltered Data: ")
print(filtered_data)
```
