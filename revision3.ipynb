```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = {
    'Days_Active': [10, 20, 30, 40, 50],
    'Reels_Posted': [2, 5, 8, 12, 15],
    'Followers_Gained': [100, 250, 400, 600, 750]
}

df = pd.DataFrame(data)

# Create a Correlation Heatmap (sns.heatmap) for this dataframe to mathematically 
# prove which column is the biggest "bestie" to Followers_Gained. (Don't forget 
# annot=True so we can see the numbers!)
matrix = df.corr()

sns.heatmap(matrix, cmap="coolwarm", vmin=-1, vmax=1, linewidths=1, linecolor='black', annot=True)
plt.title("Correlation Map")
plt.show()
```

```python
# Set your X to be Reels_Posted and your y to be Followers_Gained.
X = df[['Reels_Posted']]
y= df['Followers_Gained']

# Train a LinearRegression() model on this data.
model = LinearRegression()
model.fit(X, y)

# Draw a Scatter Plot showing the original data points (blue), and draw 
# the AI's Line of Best Fit right over it (red).
plt.scatter(X, y, color='blue', label='Actual Followers Gained')

predictions = model.predict(X)
plt.plot(X, predictions, color="red", linewidth=2, label="Best Fit Line")
plt.title("AI followers analysis")
plt.xlabel("Reels Posted")
plt.ylabel("Followers Gained")
plt.show()
```
