```python
# generate a Box Plot for a dataset and identify the outliers using the IQR method

import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Scores': [5, 7, 8, 12, 15, 18, 19, 20, 35, 37]
}

df = pd.DataFrame(data)

# A Box Plot draws a literal box around where most of the normal data lives. 
# Anything sitting way outside the box is an outlier
plt.boxplot(df['Scores'])
plt.title("Spotting the outliers")
plt.ylabel("Data Values")
plt.show()

# Finding the outliers using Interquartile range
Q1 = df['Scores'].quantile(0.25) # bottom 25%
Q3 = df['Scores'].quantile(0.75) # top 75%
IQR = Q3 - Q1 # size of the box

# anything below lower bound or above the upper bound gets kicked out
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# filtering the table
outliers = df[(df['Scores'] < lower_bound) | (df['Scores'] > upper_bound)]

print(f"\nThe mathematically proven outliers are: {outliers}")

```
