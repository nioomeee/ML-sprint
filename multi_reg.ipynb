```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv('multireg.csv')
print("1. Multiple clues regression file loaded!")
print(df.head(3))

X = df[['Area', 'Bedrooms']]
y = df['Price']

model = LinearRegression()
model.fit(X, y)
print("\n2. Model trained succesfully")

new_house = [[2000, 3]]

predicted_price = model.predict(new_house)
print(f"\nPredicted price of a 2000 sq ft house with 3 bedrooms: ${predicted_price[0]}")
```
