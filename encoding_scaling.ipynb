```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

data = {
    'Name': ['Alice', 'Charlie', 'Diana', 'Eve'],
    'Department': ['HR', 'IT', 'Finance', 'IT'],
    'Salary': [50000, 80000, 120000, 45000]
}

df = pd.DataFrame(data)
print("Raw Data")
print(df)

# Use LabelEncoder() to translate the 'Department' column from text into numbers.
encoder = LabelEncoder()
df['Department'] = encoder.fit_transform(df['Department'])
print("\nAfter encoding department:")
print(df)

# Use MinMaxScaler() to squish the 'Salary' column so all those massive numbers s
# it perfectly between 0 and 1.
scaler = MinMaxScaler()
df['Salary'] = scaler.fit_transform(df[['Salary']]) #remember [[]] for scaling!!!

print("\nAfter encoding and sclaing: ")
print(df)
```
