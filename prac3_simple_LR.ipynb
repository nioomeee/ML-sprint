```python
import pandas as pd
from sklearn.linear_model import LinearRegression

data = {
    'Sizes': [1100, 1300, 1500, 1700, 1900],
    'Prices': [210000, 250000, 290000, 330000, 370000]
}

df = pd.DataFrame(data)
model = LinearRegression()

X = df[['Sizes']]

y = df['Prices']

# Train a LinearRegression() model using this data.
model.fit(X, y)
print("Model trained successfully!")

# Predict the price for a house that is 1600 sq ft.
new_house = [[1600]]
predicted_price = model.predict(new_house)

print(f"\nPredicted value of a house of 1600 sq ft: ${predicted_price[0]}")
```
