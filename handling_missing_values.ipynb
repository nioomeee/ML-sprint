```python
import pandas as pd

data = {
    'Name': ['John', 'Anna', 'Peter', None], 
    'Age': [24, 13, None, 33]
}

df = pd.DataFrame(data)

print("\n Original messy table: ")
print(df)

# if any row has blank space, delete entire row
df_dropped = df.dropna()

print("\n After using dropna: ")
print(df_dropped)
```

```python
# instead of deleting, let's fill with mean
df_patched = df.copy()

# Using mean for numbers
df_patched['Age'] = df_patched['Age'].fillna(df['Age'].mean())

# Using "Unknown" for names
df_patched['Name'] = df_patched['Name'].fillna("Unknown")

print("\n After using fillna: ")
print(df_patched)
```
