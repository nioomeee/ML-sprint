```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

data = {
    'Credit_Score': [650, 700, 720, 680, 750, 640],
    'Age': [25, 30, 45, 35, 50, 28],
    'Income': [3, 5, 8, 4, 10, 3],
    'Loan_Approved': ['No', 'Yes', 'Yes', 'No', 'Yes', 'No']
}

df = pd.DataFrame(data)
print("\nOriginal data: ")
print(df)

encoder = LabelEncoder()

df['Loan_Approved'] = encoder.fit_transform(df['Loan_Approved'])

X = df[['Credit_Score', 'Age', 'Income']]

y = df['Loan_Approved']

model = DecisionTreeClassifier()

model.fit(X, y)

print("\n Decision tree has been trained successfully")
```

```python

# fixing dimensions of the canvas
plt.figure(figsize=(10, 8))

plot_tree(model, feature_names=['Credit_Score', 'Age', 'Income'], class_names=['Yes', 'No'], filled=True, rounded=True)

plt.title("Decision Tree")
plt.show()
```
