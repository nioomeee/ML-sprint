```python
import pandas as pd
from sklearn.linear_model import LinearRegression

df = pd.read_csv('house_data.csv')
print("1. CSV file read successfully!")
print(df.head(3))

X = df[['Size']]
y = df['Price']

model = LinearRegression()
model.fit(X, y)
print("\n2. Model trained succesfully!")

new_house = [[3000]]
prediction = model.predict(new_house)

print("\n Pridicted price for a 300 sq feet house: $", prediction[0])
```
