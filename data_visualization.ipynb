```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    'Emp_ID': [1, 2, 3, 4, 5, 6, 3, 8],
    'Age': [25, 30, np.nan, 28, 35, 30, np.nan, 40],
    'Department': ['HR', 'IT', 'IT', 'Finance', 'HR', 'IT', 'IT', 'Finance'],
    'Experience': [2, 5, 4, np.nan, 10, 5, 4, 15],
    'Salary': [30000, 50000, 45000, 40000, 70000, 50000, 45000, 90000]
}

df = pd.DataFrame(data)
print("Original dataframe: ")
print(df)

# histogram to see how salaries are spread out
plt.hist(df['Salary'], bins=4, color='lightpink', edgecolor='blue')
plt.title('\n Distribution of employee salaries: ')
plt.xlabel("Salary Range")
plt.ylabel("Number of Employees")

plt.show()

# Plotting a scatter plot to show relationship b/w experience and salary
plt.scatter(df['Experience'], df['Salary'], color='purple')
plt.title("Experience vs Salary")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.show()
```
