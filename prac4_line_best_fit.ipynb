```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = {
    'Sizes': [1100, 1300, 1500, 1700, 1900],
    'Prices': [210000, 250000, 290000, 330000, 370000]
}

df = pd.DataFrame(data)
print("Original data:")
print(df)

X = df[['Sizes']]
y = df['Prices']

model = LinearRegression()
model.fit(X, y)
print("\nModel trained succesfully")

plt.scatter(X, y, color='purple', label='Actual Prices')

predictions = model.predict(X)
plt.plot(X, predictions, color='black', linewidth=2, label='Line of best fit')

plt.title('AI trend analysis')
plt.xlabel("Size (sq ft)")
plt.ylabel("Price ($)")
plt.show()

```
