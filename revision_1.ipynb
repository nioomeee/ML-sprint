```python
import pandas as pd
import numpy as np
import re
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize

def preprocess(text):
    clean = re.sub(r'<.*?>', '', text)
    return clean

data = {
    'Join_Date': ['12/01/2023', '2023.05.05', '03-10-2023'],
    'Feedback': ['<html>Loved it!</html>', '<b>Worst app</b>', 'It was okay'],
    'Rating': [9, np.nan, 5],
    'Raw_Words': ['running', 'better', 'cats']
}
df = pd.DataFrame(data)
print("Original data")
print(df)

# Convert the Join_Date column to a standard datetime format
df['Join_Date'] = pd.to_datetime(df['Join_Date'], format='mixed')
print("\n After datetime conversion: ")
print(df)

# Write a quick regex command to strip the HTML tags (<.*?>) out of the Feedback column.
df['Feedback'] = df['Feedback'].apply(preprocess)
print("\n After removing HTML tags: ")
print(df)

# Fill the missing NaN value in the Rating column with the mean (average) of the other ratings.
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())
print("\n After filling out the missing value in Ratings column with mean: ")
print(df)

# Use WordNetLemmatizer() to find the root words for the Raw_Words column (you can just apply it to the list of words).
lemmatizer = WordNetLemmatizer()
df['Raw_Words'] = df['Raw_Words'].apply(lemmatizer.lemmatize)
print("\n After Lemmatization: ")
print(df)
```
