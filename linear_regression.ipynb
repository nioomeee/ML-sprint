```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

data = {
    'House_Size': [1000, 1500, 2000, 2500, 3000],
    'House_Price': [200000, 250000, 300000, 350000, 400000]
}
df = pd.DataFrame(data)

# X must be having double brackets
X = df[['House_Size']]

# y must have only 1 bracket
y = df['House_Price']

model = LinearRegression()

model.fit(X, y)

print("\n Linear Regression model trained!")

new_house = [[1800]]

predicted_price = model.predict(new_house)

print("\n Predicted price of the new 1800 sq ft house -> $",predicted_price[0])
```
