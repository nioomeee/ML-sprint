```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder

data = {
    'Credit_Score': [650, 700, 720, 680, 750, 640],
    'Age': [25, 30, 45, 35, 50, 28],
    'Income': [3, 5, 8, 4, 10, 3],
    'Loan_Approved': ['No', 'Yes', 'Yes', 'No', 'Yes', 'No']
}

df = pd.DataFrame(data)
print("\nOriginal data: ")
print(df)

# label encoder turns Yes and No to 1 and 0
encoder = LabelEncoder()
df['Loan_Approved'] = encoder.fit_transform(df['Loan_Approved'])

# X is clues and y is answers
# The input field the model uses to make predictions
X = df[['Credit_Score', 'Age', 'Income']]

# the final answer it needs to predict
y  = df['Loan_Approved']

# inititialize decision tree
model = DecisionTreeClassifier()
model.fit(X, y)

print("Decision Tree has finished studying and is fully trained!")
```

```python
# New person data -> credit score = 700, Age = 30, Income = 8
new_person = [[700, 30, 8]]

# AI takes a guess
numeric_prediction = model.predict(new_person)

# answer was in 0 or 1, we need to convert it back to yes or no using encoder
final_answer = encoder.inverse_transform(numeric_prediction)

print("\n Will the loan be approved for this new person? -> ", final_answer[0])
```
