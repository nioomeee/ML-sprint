```python
# turn a dictionary of data into a DataFrame and clean it up
import pandas as pd 
import numpy as np 

data = {
    'Emp_ID': [1, 2, 3, 4, 5, 6, 3, 8],
    'Age': [25, 30, np.nan, 28, 35, 30, np.nan, 40],
    'Department': ['HR', 'IT', 'IT', 'Finance', 'HR', 'IT', 'IT', 'Finance'],
    'Experience': [2, 5, 4, np.nan, 10, 5, 4, 15],
    'Salary': [30000, 50000, 45000, 40000, 70000, 50000, 45000, 90000],
}

df = pd.DataFrame(data)

print("Dataframe with duplicates: ")
print(df)

df = df.drop_duplicates()
print("\n After removing duplicates: ")
print(df)

print("\n Missing values before cleaning: ")
print(df.isnull().sum())

# Replacing missinf values with their mean
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Experience'] = df['Experience'].fillna(df['Experience'].mean())

print("\n After cleaning: ")
print(df)

# Grabbing specific rows
first_person = df.iloc[0]
print("\n Details of the 1st person: ")
print(first_person)

# Grabbing range of rows
first_two_people = df.iloc[0:2]
print("\n Details of the 1st two people")
print(first_two_people)
```
