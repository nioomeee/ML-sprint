```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {
    'Fruits': ['Apple', 'Banana', 'Orange', 'Apple', 'Banana', 'Banana', 'Orange', 'Apple']
}

df = pd.DataFrame(data)
print("Original data: ")
print(df)

# .value_counts() counts how many times each fruit exists
fruit_counts = df['Fruits'].value_counts()
```

```python
# Barchart (X, Y)
# Fruit_counts.index is names (Apple, Banana, Orange), 
# Fruit_counts.values is numbers (3, 3, 2)

plt.bar(fruit_counts.index, fruit_counts.values, color=["Red", "Yellow", "Orange"])

plt.title("Fruit Frequency")
plt.xlabel("Fruit Name")
plt.ylabel("Fruit Frequency")

plt.show()
```

```python
# Heatmap
data_features = {
    'Feature_1': [10, 20, 30, 40, 50],
    'Feature_2': [50, 40, 30, 20, 10],
    'Feature_3': [12, 18, 29, 35, 60]
}

df_features = pd.DataFrame(data_features)

# Step 1: Calculate the math
# .corr() checks if Feature 1 goes up, feature 2 goes up too?

matrix = df_features.corr()

# Step 2: Paint the heatmap
# annot = True puts actual numbers inside colored boxes
# coolwarm is red to blue color palette
# vmin = -1 -> deep red, vmax = 1 -> dark blue 
sns.heatmap(matrix, annot=True, cmap="coolwarm", vmin=-1, vmax = 1, linewidths = 1, linecolor='black')
plt.title("Feature Correlation Heatmap")
plt.show()
```
