```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 1. Data. Just the X, there's no Y
data = np.array([
    [185, 72], [170, 56], [168, 60], [179, 68], [182, 72], [188, 77], [180, 71], [180, 70]
])

# 2.calling the KMeans AI 
kmeans = KMeans(n_clusters=2, random_state=42)

# 3. train it
kmeans.fit(data)

# 4. Get results
labels = kmeans.labels_ # tells us which group(0/1) everyone got assigned to
centroids = kmeans.cluster_centers_ # exact coordinates of centroids

print(f"\n Assigned groups: {labels}")
print(f"\n Centroid coordinates: \n {centroids}")

# Visualization !!!!!!!!!!!!
plt.figure(figsize=(10, 8))

# plotting regular points, colored by group
plt.scatter(data[:, 0], data[:, 1], cmap="coolwarm", c=labels, s=100, label="Data Points")

# plotting the centroid
plt.scatter(centroids[:, 0], centroids[:, 1], c="gold", marker="*", s=300, edgecolors='black', label="Centroids")

plt.xlabel("Height (cm)")
plt.ylabel("Weight (kg)")
plt.title("K-Means Clustering (K=2)")
plt.legend()
plt.show()
```
